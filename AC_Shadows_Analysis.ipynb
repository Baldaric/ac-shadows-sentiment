# Import Libaries

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
from collections import Counter
```

# Data Loading

Data we are using are obtained from steam API. From there we obtained around 6000 data. The data is then labeled using BERT and analyzed further. For now we only filter the reviews that are in English language.

```python
df = pd.read_csv('AC_Shadows_Bert.csv')
```

```python
df.drop(columns="Unnamed: 0",inplace=True )
```

```python
df.head()
```

# Distribution

```python
sentiment_counts = df["bert_sentiment"].value_counts()

plt.figure(figsize=(6,4))
sentiment_counts.plot(kind="bar", color=["green", "red", "grey"])
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

Based on this distribution, the reviews are mostly positive indicates that the game are liked by the players.

# Playtime

```python
# Create a boxplot of sentiment scores by playtime
sns.boxplot(x='bert_sentiment', y='author_playtime_forever', data=df)
plt.yscale('log')  # Makes visualization clearer for large numbers
plt.title("Playtime Distribution by Sentiment")
plt.show()
```

### 🔍 Analysis of Playtime Distribution by Sentiment:
 - The box plot shows the distribution of playtime (author_playtime_forever) categorized by three sentiment groups: Neutral, Negative, and Positive.
 - The playtime is plotted on a logarithmic scale, ranging from around 10 (low playtime) to 10,000 (high playtime).

### 📊 Key Observations:
Median Playtime:
 - The Negative sentiment group has a median playtime slightly higher or comparable to the Neutral and Positive groups.
 - This suggests that players with Negative reviews have spent a significant amount of time playing the game, which means their criticisms are likely based on extensive experience with the game.

#### Range & Outliers:
 - Negative reviews also show a wide range of playtime, with numerous outliers representing very high playtime.
 - This further supports the idea that negative feedback may come from dedicated players who have spent substantial time with the game and are invested enough to criticize its flaws.

#### Genuine Criticism:
 - Players who spend more time on a game and leave negative reviews are likely providing constructive criticism rather than casual, surface-level complaints.
 - Their detailed feedback can be valuable for identifying core issues with the game.

### ✅ Conclusion:
 - Negative reviews are more likely to be genuine or detailed because they tend to come from players who have spent considerable time with the game.
 - Positive reviews may be less critical if they come from casual players who only experienced the game superficially.
 - Though the positive reviews also indicates that most of the players already satisfied with the game.

# Time

```python
# Step 1: Convert timestamp to datetime
df['timestamp_created'] = pd.to_datetime(df['timestamp_created'], unit='s')

# Step 2: Create a 'date' column (grouping by day)
df['review_date'] = df['timestamp_created'].dt.date

# Step 3: Count sentiments over time
trend = df.groupby(['review_date', 'bert_sentiment']).size().unstack().fillna(0)

# Optional: Use rolling average (e.g., 3-day window) for smoother trend
trend_rolling = trend.rolling(window=3).mean()

# Step 4: Plot
plt.figure(figsize=(10, 5))
trend_rolling.plot(color=['#E63946', '#ADB5BD','#6DD47E'], linewidth=2)

plt.xlabel("Date")
plt.ylabel("Number of Reviews")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

### Key Insights
 - Positive reviews peaked at launch (~March 22) but dropped sharply after a few days.
 - This suggests strong initial interest, possibly driven by marketing, visuals, or fan expectations.
 - The steep decline indicates that early excitement didn’t sustain, possibly due to unmet expectations or gameplay flaws.
 - Negative reviews remained steady for a longer period, indicating ongoing player frustrations.
 - These may stem from bugs, performance issues, or disappointment with deeper gameplay elements discovered post-launch.
 - Overall review activity dropped significantly after just 2 weeks.
 - This could mean the game lacked long-term retention, replay value, or community engagement to keep players invested.
